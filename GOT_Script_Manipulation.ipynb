```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'/content/conv.txt', sep=": ", header=None, names=['name','dialogue'])
df = df.set_index('name')
df
```

```python
df_grouped = df.groupby(['name']).agg(lambda col: ''.join(pd.unique(col)))
df_grouped
```

```python
# 1. No of Unique Characters
len(df_grouped)
```

```python
# 2. Create a new text file for each of the characters with their name and store 
# the unique words said by them in their respective file. 
# Store one word in one line.
i=0
for index, row in df_grouped.iterrows():
    if i > len(df_grouped):
       break
    else:
       f = open(str(index)+'.txt', 'w')
       f.write('\n'.join(row[0].split(' ')))
       f.close()
       i+=1
```
